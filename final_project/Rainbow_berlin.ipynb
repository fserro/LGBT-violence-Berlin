```python
import numpy as np
from PIL import Image
```

```python
bear = Image.open("icons/berlin_bear.png")
```

```python
berlin_array = np.zeros(shape=(420, 800, 3), dtype='int8')
```

```python
berlin_array[:,:,:] = 255
berlin_array[:69,:,] = (228, 3, 3)
berlin_array[70:139,:,] = (255, 140, 0)
berlin_array[140:209,:,] = (255, 237, 0)
berlin_array[210:279,:,] = (0, 128, 38)
berlin_array[280:349,:,] = (0, 77, 255)
berlin_array[350:420,:,] = (117, 7, 135)




```

```python
berlin_flag = Image.fromarray(berlin_array, 'RGB')
berlin_flag
```

```python
bear_small = bear.resize(size=(182, 300) )
```

```python
bear_small
```

```python
berlin_flag.paste(bear_small, (300, 54), bear_small)
berlin_flag
```
